```python
import pandas as pd ;
import matplotlib.pyplot as plt ;
```

<p1> The following code assigns to the variable 'file_name', the name of the csv file containing the global covid-19 data.

```python
file_name='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv' ;
```

```python
mdf=pd.read_csv(file_name);
```

The following code creates dataframe slices corresponding to each country(rows are filtered out for each country)

```python
india=mdf[mdf['Country/Region']=='India'];
russia=mdf[mdf['Country/Region']=='Russia'];
canada=mdf[mdf['Country/Region']=='Canada'];
china=mdf[mdf['Country/Region']=='China'];
brazil=mdf[mdf['Country/Region']=='Brazil'];
italy=mdf[mdf['Country/Region']=='Italy'];
```

The following code consolidates the data of each country(in case of multiple rows)

```python
#i=list(india.iloc[:,4:].sum(axis=0));
i=list(india.iloc[:,list(range(50,len(india.columns),7))].sum(axis=0));
c=list(canada.iloc[:,list(range(50,len(canada.columns),7))].sum(axis=0));
r=list(russia.iloc[:,list(range(50,len(russia.columns),7))].sum(axis=0));
ch=list(china.iloc[:,list(range(50,len(china.columns),7))].sum(axis=0));
b=list(brazil.iloc[:,list(range(50,len(brazil.columns),7))].sum(axis=0));
it=list(italy.iloc[:,list(range(50,len(italy.columns),7))].sum(axis=0));
```

The following code defines the x axis of the plot. The x axis is composed of the 'dates' of the time series.

```python
x=list(mdf.columns)[50:len(india.columns):7];
#x=list(mdf.columns)[4:] ;
```

```python
%matplotlib inline 
plt.figure(figsize=(18,9)) ;
plt.plot(x,i,'r',linewidth=2.0) ;
plt.plot(x,c,'g',linewidth=2.0) ;
plt.plot(x,r,'b',linewidth=2.0) ;
plt.plot(x,ch,'c',linewidth=2.0) ;
plt.plot(x,b,'k',linewidth=2.0) ;
plt.plot(x,it,'m',linewidth=2.0) ;
#plt.title('Covid graph of India') ;
plt.xticks(rotation=90);
plt.ylabel('Covid+ cases - cumulative') ;
plt.legend(['India','Canada','Russia','China','Brazil','Italy'], loc='upper left') ;
plt.show() ;
```
